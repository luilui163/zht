```python
import urllib2
import json

url=r'http://query.sse.com.cn/infodisplay/queryLatestBulletinNew.do?&jsonCallBack=jsonpCallback62474&productId=&reportType2=DQGG&reportType=ALL&beginDate=&endDate=&pageHelp.pageSize=25&pageHelp.pageCount=50&pageHelp.pageNo=100&pageHelp.beginPage=1&pageHelp.cacheSize=1&pageHelp.endPage=11&_=1469089651088'


referer='http://www.sse.com.cn/disclosure/listedinfo/regular/'
header={'referer':referer}
req=urllib2.Request(url,headers=header)
content=urllib2.urlopen(req).read()
```

```python
content
```

```python
d=content.split('(')[1][:-1]
```

```python
js=json.loads(d)
```
